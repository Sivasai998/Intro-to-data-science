# data 4
https://catalog.data.gov/dataset/department-of-state-business-filings-beginning-1991

```python
df4 = pandas.read_csv('https://data.ny.gov/api/views/m7i3-tv6j/rows.csv?accessType=DOWNLOAD')
df4.shape
```

```python
df4.head()
```

### Is this data tidy or not?

(Yes or no)

*if the data is not tidy, then convert the data to a tidy dataframe*


No

There are 24 column variables,

All months can be displayed under one variable(month) and its values can be displayed under Number of certifications.

```python
#melting
df4 = pandas.melt(df4,['Full Certificate Code', 'Certificate Type Code',
       'Certificate Type Description', 'Authority Type Code',
       'Authority Type Description', 'Business Type Code',
       'Business Type Description', 'Name Type Code', 'Name Type Description',
       'Description', 'Year','Total'], var_name="Month", value_name="Number of certifications")
df4 = df4.sort_values(by=["Month"])
df4.head()
```

```python
#shape
#there are 24 columns present in the starting after changing it consists of 14 columns.
df4.shape
```

```python
#printing columns varaiables
df4.columns
```
