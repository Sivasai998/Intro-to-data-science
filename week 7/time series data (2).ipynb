Create two bar graphs of the power consumption per hour.

One bar graph has 24 bars; each bar is the average across all days for that hour​

Submit the .ipynb file containing the analysis and the generated pictures.​

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import time
%matplotlib inline
data1=pd.read_csv(r"C:\Users\nvgsiva sai\Downloads\RollingSystemDemand_1.csv",index_col=False,skiprows=1,skipfooter=1,engine='python',header=None)
data2=pd.read_csv(r"C:\Users\nvgsiva sai\Downloads\RollingSystemDemand_2.csv",index_col=False,skiprows=1,skipfooter=1,engine='python',header=None)
data3=pd.read_csv(r"C:\Users\nvgsiva sai\Downloads\RollingSystemDemand_3.csv",index_col=False,skiprows=1,skipfooter=1,engine='python',header=None)
data=pd.concat([data1,data2,data3],axis=0)
data.columns=["VD",'date_time',"Value"]
data["date_time"]=pd.to_datetime(data['date_time'].astype(str),format = '%Y%m%d%H%M%S')
df = pd.DatetimeIndex(data['date_time'])
hour = data.groupby([df.hour])['Value'].mean()
hour1 = data.groupby([df.date,df.hour])['Value'].sum()
fig,axes=plt.subplots(1,2,figsize=(15,5))
hour.plot.bar(ax=axes[0])
hour1.plot.bar(ax=axes[1])
axes[0].set_title('Power consumption per hour')
axes[0].set_ylabel('Value')
axes[1].set_title('Power Consumption per hour')



```
